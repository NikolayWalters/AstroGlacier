```python
import lightkurve as lk
import numpy as np
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle
```

```python
tpf = lk.search_targetpixelfile('GD356', mission='TESS').download(quality_bitmask='default')
# Create a custom aperture matrix or grid
custom_aperture = np.zeros((tpf.shape[1], tpf.shape[2]), dtype=bool)

# Define the x and y coordinates for the center of your aperture
x_center = 10
y_center = 10

# Define the size of your aperture
aperture_size = 1

# Define the shape of the aperture 
aperture_shape = 'square' # or 'circle'

if aperture_shape == 'square':
    custom_aperture[y_center-aperture_size:y_center+aperture_size, x_center-aperture_size:x_center+aperture_size] = 1
    
tpf.plot(aperture_mask=custom_aperture, mask_color='red')
plt.show()

```

```python
lc = tpf.to_lightcurve(aperture_mask=custom_aperture.astype(int))
lc.plot()
```

```python
lc1 = lc
freq = np.linspace(0.001,10,10000) 
powerPl1 = LombScargle(lc1.time.value,lc1.flux.value).power(freq)
powerPl1 = powerPl1/max(powerPl1)
plt.figure(figsize=(20,10))
plt.plot(freq, powerPl1)
```

```python
search_result = lk.search_lightcurve('NLTT 5306', mission='TESS')
search_result
```

```python
lc = search_result[0].download()
```

```python
plt.plot(lc.time.value, lc.flux.value, 'k,')
f1 = lc.flux.value
nan_ind = np.isnan(f1)
f1 = np.delete(f1, nan_ind)
t1 = lc.time.value
t1 = np.delete(t1, nan_ind)
f1 = f1/np.mean(f1)
plt.show()
plt.plot(t1, f1, 'k,')
```

```python
lc = search_result[2].download()
plt.plot(lc.time.value, lc.flux.value, 'k,')
f2 = lc.flux.value
nan_ind = np.isnan(f2)
f2 = np.delete(f2, nan_ind)
t2 = lc.time.value
t2 = np.delete(t2, nan_ind)
f2 = f2/np.mean(f2)
plt.show()
plt.plot(t2, f2, 'k,')
```

```python
freq = np.linspace(0.001,20,10000) 
allF = np.concatenate((f1,f2))
allT = np.concatenate((t1,t2))
powerPl = LombScargle(allT, allF).power(freq)
plt.figure(figsize=(20,10))
plt.plot(freq, powerPl)
plt.plot([1/(1.7/24),1/(1.7/24)],[0,0.001])
```

```python
(1/freq[np.argmax(powerPl)])*24*60
```
